```python
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/hayase/study-analystics-ai/data/sample.csv")
```

```python
df
```

```python
# subjectの日本語化
mapping = {
    "応用情報": "Applied Information",
    "開発": "Development",
}
df["subject"] = df["subject"].replace(mapping)
df.head(3)
```

```python
df.describe()
```

```python
# 勉強時間と理解度の関係
df.plot(kind = 'scatter', x = 'study_time', y = 'understanding')
```

study_timeとunderstandingには正の相関がありそうである。

```python
# subjectごとの勉強時間の合計
subject_sum = df.groupby("subject")["study_time"].sum()
subject_sum.plot(kind = 'bar')

subject_sum.plot(
    kind = 'bar',
    title="Study time by each subject",
    ylabel="Hours",
    xlabel="Subject"
)
```

相対的にMLの勉強時間が多く、開発の時間が少ない。勉強時間の見直しが必要である。

```python
df["date"] = pd.to_datetime(df["date"])
daily_sum = df.groupby("date")["study_time"].sum()

daily_sum.plot(
    title="Daily Study Time",
    ylabel="Hours",
    xlabel="Date"
)
```
